```python
# !pip install opencv-python
import cv2
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') # We load the cascade for the face.
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml') # We load the cascade for the eyes.
```

```python
import numpy as np
snobel_filter = np.array([[2,1,5],
                  [3,5,4],
                  [4,1,2]])

def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    count = len(faces)
    
    for (x, y, w, h) in faces: # For each detected face:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
       
        cv2.putText(frame,"total faces:{}".format(count),(10,20),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
        
        roi_gray = gray[y:y+h, x:x+w]
        
        frame[y:y+h, x:x+w]=cv2.filter2D(frame[y:y+h, x:x+w],-1,snobel_filter)
#         eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 3)
#         count1 = len(eyes)
#         for (ex, ey, ew, eh) in eyes:
#             cv2.rectangle(roi_color,(ex, ey),(ex+ew, ey+eh),(0,255, 0), 2)
#             cv2.putText(frame,"total eyes:{}".format(count1),(10,40),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
            
        
    return frame
```

```python
video_capture = cv2.VideoCapture(0) # We turn the webcam on.
print(video_capture.isOpened())
while video_capture.isOpened():
    ret, frame = video_capture.read() # We get the last frame.
    print(ret)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame) 
    cv2.imshow('Video', canvas)
    
    if cv2.waitKey(1) & 0xFF == ord('e'):
        break # We stop the loop.
video_capture.release() 
cv2.destroyAllWindows()
```
